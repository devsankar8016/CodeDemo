```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the dataset
df = pd.read_csv("/content/drive/MyDrive/Almabetter Drive/Almabetter_Notes/netflix_titles.csv")

# 1. Identify missing values
# Check the number of missing values in each column
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)

#df.head()
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)

# Remove irrelevant columns
df.drop(['show_id', 'description'], axis=1, inplace=True)
df.head()
```

```python
import matplotlib.pyplot as plt

# Bar chart: Number of movies vs TV shows
df['type'].value_counts().plot(kind='bar')
plt.title('Number of Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()

# Line chart: Trend in number of new titles added each year
df['release_year'].value_counts().sort_index().plot(kind='line')
plt.title('Trend in Number of New Titles Added Each Year on Netflix')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
# Extract genres from the "listed_in" column
genres = df['listed_in'].str.split(', ')
all_genres = [genre for sublist in genres for genre in sublist]

# Count the occurrences of each genre
genre_counts = pd.Series(all_genres).value_counts()

# Plot the top 10 genres
top_10_genres = genre_counts.head(10)
plt.figure(figsize=(10, 6))
top_10_genres.plot(kind='bar')
plt.title('Top 10 Most Popular Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
# Write your code here
import seaborn as sns

# Load the dataset
iris_df = sns.load_dataset('iris')
```

```python
# Display the first few rows of the dataset
print(iris_df.head())

# Display information about the dataset
print(iris_df.info())

# Statistical summary of the dataset
print(iris_df.describe())
```

```python
import matplotlib.pyplot as plt

# Scatter plot: Sepal length vs Sepal width
sns.scatterplot(data=iris_df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width')
plt.show()

# Box plot: Petal length by Species
sns.boxplot(data=iris_df, x='species', y='petal_length')
plt.title('Petal Length by Species')
plt.show()

# Violin plot: Petal width by Species
sns.violinplot(data=iris_df, x='species', y='petal_width')
plt.title('Petal Width by Species')
plt.show()

# Pair plot: Pairwise relationships between variables
sns.pairplot(data=iris_df, hue='species')
plt.title('Pairwise Relationships')
plt.show()

# Heatmap: Correlation matrix
corr_matrix = iris_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Customized scatter plot: Sepal length vs Sepal width
sns.scatterplot(data=iris_df, x='sepal_length', y='sepal_width', hue='species', palette='Set2')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()

# Customized box plot: Petal length by Species
sns.boxplot(data=iris_df, x='species', y='petal_length', palette='Set3')
plt.title('Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.show()

# Customized violin plot: Petal width by Species
sns.violinplot(data=iris_df, x='species', y='petal_width', palette='Set1')
plt.title('Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width')
plt.show()
```
